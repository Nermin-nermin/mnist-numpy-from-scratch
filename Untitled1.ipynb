```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import gradio as gr
```

```python
train_df = pd.read_csv('/content/training.csv', encoding='ISO-8859-1')
val_df = pd.read_csv('validation-1.csv')
test_df = pd.read_csv('test.csv')

for df_name, df in [('train', train_df), ('validation', val_df), ('test', test_df)]:
  assert 'text' in df.columns, f'{df_name} dataframe is missing the "text" column'
  assert 'label' in df.columns, f'{df_name} dataframe is missing the "label" column'
  df['text'] = df['text'].fillna('')

train_df.head()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
port_stem = PorterStemmer()
```

```python
stop_words = set(stopwords.words('english'))
for word in ['not', 'no', 'nor']:
  if word in stop_words:
    stop_words.remove(word)

def data_processing(data):

  processed_data = re.sub('[^a-zA-Z]', ' ', data)
  processed_data = processed_data.lower()
  processed_data = processed_data.split()
  processed_data = [port_stem.stem(word) for word in processed_data if word not in stop_words]
  processed_data = ' '.join(processed_data)

  return processed_data
```

```python
train_df['processed_data'] = train_df['text'].apply(data_processing)
```

```python
train_df['processed_data'].head()
```

```python
train_df.head()
```

```python
print(train_df['label'])
```

```python
X_train = train_df['processed_data']
Y_train = train_df['label']

X_val = val_df['text'].apply(data_processing)
Y_val = val_df['label']

X_test = test_df['text'].apply(data_processing)
Y_test = test_df['label']
```

```python
print(X_train.shape, X_val.shape, X_test.shape)
```

```python
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))

X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
X_val = vectorizer.transform(X_val)
```

```python
model = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score on the training data: ', training_data_accuracy)
```

```python
X_val_prediction = model.predict(X_val)
val_data_accuracy = accuracy_score(X_val_prediction, Y_val)

from sklearn.metrics import classification_report
print("Validation Classification Report:")
print(classification_report(Y_val, X_val_prediction))

from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(Y_val, X_val_prediction)
plt.show()
```

```python
print('Accuracy score on the validation data: ', val_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy score on the test data: ', test_data_accuracy)
```

```python
def predict_sentiment(text):
    processed_text = data_processing(text)
    vectorized_text = vectorizer.transform([processed_text])
    return model.predict(vectorized_text)[0]

interface = gr.Interface(fn=predict_sentiment,
                      inputs=gr.Textbox(lines=2, placeholder='Enter your text here: '),
                      outputs='text',
                      title='Sentiment Analysis')

interface.launch(share=True, debug=True)
```
